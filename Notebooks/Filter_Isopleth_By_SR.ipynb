# Filter Isopleth Data by Solar Radiation

Join isopleth ozone data with solar radiation observations and remove rows below a configurable `SR_THRESHOLD`.

```python
from pathlib import Path

import pandas as pd

ISOPLETH_PATH = Path("../data/utah/raw_data/HW_data_winter.csv")
SR_PATH = Path("../data/SR_Data/SR_Data_HW.csv")
SR_THRESHOLD = 710  # W/m^2; adjust as needed
```

```python
isopleth_df = pd.read_csv(ISOPLETH_PATH, parse_dates=["dt"])

sr_df = (
    pd.read_csv(SR_PATH, index_col=0, parse_dates=True)
    .rename_axis("dt")
    .reset_index()
)

isopleth_df.head()
```

```python
merged_df = isopleth_df.merge(
    sr_df,
    on="dt",
    how="inner",
    # validate="one_to_one",
)

filtered_df = merged_df[merged_df["solar_radiation_w_m2"].ge(SR_THRESHOLD)].copy()
filtered_df.head()
```

```python
print(
    f"Rows retained: {len(filtered_df)} of {len(merged_df)} (threshold >= {SR_THRESHOLD} W/m^2)"
)
filtered_df.tail()
```

```python
filtered_df.to_csv("../data/utah/filtered_by_SR_data/filtered_non-max_winter",index=False)
```
