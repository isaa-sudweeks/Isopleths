# Weekend vs Weekday Data Split

This notebook splits the dataset into weekend and weekday data based on the days of the week.

```python
import pandas as pd
import os
```

```python
# Load the data
file_path = "/Users/isaacsudweeks/Desktop/Work/Improved_Isopleths/data/utah/filtered_by_clearing_index/SR_clearing_nonmax_summer_1000.csv"
df = pd.read_csv(file_path)

# Display the first few rows
df.head()
```

```python
# Convert 'dt' column to datetime
df['dt'] = pd.to_datetime(df['dt'])

# Extract day of week (0=Monday, 6=Sunday)
df['day_of_week'] = df['dt'].dt.dayofweek

df[['dt', 'day_of_week']].head()
```

```python
# Split into weekday and weekend
# Weekday: 0-4 (Mon-Fri)
# Weekend: 5-6 (Sat-Sun)

weekday_df = df[df['day_of_week'] <= 4].copy()
weekend_df = df[df['day_of_week'] >= 5].copy()

print(f"Total records: {len(df)}")
print(f"Weekday records: {len(weekday_df)}")
print(f"Weekend records: {len(weekend_df)}")
```

```python
# Verify the split
print("Weekday days present:", weekday_df['day_of_week'].unique())
print("Weekend days present:", weekend_df['day_of_week'].unique())
```

```python
# Save the split datasets
weekday_df.to_csv("../data/utah/filtered_by_clearing_index/Weekend_Split/SR_clearing_nonmax_summer_weekday.csv", index=False)
weekend_df.to_csv("../data/utah/filtered_by_clearing_index/Weekend_Split/SR_clearing_nonmax_summer_weekend.csv", index=False)
```
